```python
import requests
import json
```

```python
url = f"https://date.nager.at/api/v3/CountryInfo/1" #this call gets the country info given the country code. I chose 1 therefore it is at the end of the link

country_info_data = requests.get(url) #this utilizes the request function to fetch the country info data
json_file = country_info_data.json() #this puts the info into a json file

json_file #this prints out the file
```

```python
url = f"https://date.nager.at/api/v3/AvailableCountries" #this call gets all available countries
                                                        #this call has no parameters
available_countries_data = requests.get(url)
json_file = available_countries_data.json()

json_file #
```

```python
url = f"https://date.nager.at/api/v3/LongWeekend/2023/82" #this call gets long weekends given a year and country code
                                                        #I chose the year 2023 and the country code 82 which is at the end of the link
long_weekend_data = requests.get(url)
json_file = long_weekend_data.json()

json_file
```

```python
url = f"https://date.nager.at/api/v3/PublicHolidays/2023/82" #this call gets public holidays given a year and country code
                                                        #I chose the year 2023 and the country code 82 which is at the end of the link
public_holiday_data = requests.get(url)
json_file = public_holiday_data.json()

json_file
```

```python
url = f"https://date.nager.at/api/v3/IsTodayPublicHoliday/UK?offset=0" #this call checks if today is a public holiday given a country code
                                                        #I chose the country code UK which is included within the link
today_public_holiday_data = requests.get(url)
json_file = today_public_holiday_data.json()

json_file
```

```python
url = f"https://date.nager.at/api/v3/NextPublicHolidays/82" #this call gets the upcoming public holidays for the next 365 days given a country code
                                                            #I chose the year 82 which is at the end of the link
next_public_holiday_data = requests.get(url)
json_file = next_public_holiday_data.json()

json_file
```
